```python
import csv
import pandas as pd
import numpy as np
```

```python
# 提取电影的特征
filename = r"E:\VSCODE\机器学习\电影推荐系统\resources\movies.csv"
list1=[]
with open(filename, 'r',encoding='utf-8') as file:
    reader = csv.DictReader(file)
    column = [row['genres'] for row in reader]
    for i in column:
        list1.append(i.split('|'))
print(list1)
```

```python
# 去除重复特征的电影

list2=[]
for n in list1:
    if n not in list2:
        list2.append(n)
list2
```
